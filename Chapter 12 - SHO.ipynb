#Chapter 12 - Simple Harmonic Oscillator
Some of the tools for using the SHO

```python
%pylab inline
```

```python
from qutip import *
```

```python
a = destroy(20)
```

```python
n = a.dag()*a
```

```python
psi = coherent(20,2)
```

```python
psi.dag()*n*psi
```

```python
psi_dm = ket2dm(psi)
```

```python
fock_distribution(psi_dm)
```

Also look at other coherent states:

```python
fock_distribution(coherent_dm(40,4))
```

And some other density matrices, like a fock-state (or n eigenstate):

```python
fock_distribution(fock_dm(20,2))
```

And a thermal state:

```python
fock_distribution(thermal_dm(20,2))
```

##Finally, can visualize the phasor using what is called the Wigner function:

```python
xvec = linspace(-10,10,200) # Create an array for the phase space coordinates.
contourf(xvec, xvec, wigner(coherent_dm(20,2),xvec,xvec),20)
```
